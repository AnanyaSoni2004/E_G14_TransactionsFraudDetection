```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100
```

# Loading Cleaned Datasets

```python
cleaned_data_path = '../data/processed/'


def load_dataset(path):
    df = pd.read_csv(f'{cleaned_data_path}{path}')
    return df

# Load all raw CSV files
account_profiles_df_processed = load_dataset("account_profiles_clean.csv")
fraud_patterns_df_processed = load_dataset("fraud_patterns_clean.csv")
network_edges_df_processed = load_dataset("network_edges_clean.csv")
time_series_stats_df_processed = load_dataset("time_series_stats_clean.csv")
transactions_df_processed = load_dataset("transactions_clean.csv")

print("All cleaned CSV files loaded.")
```

# Prepare Data for Statistical Analysis

```python
# For statistical analysis, we primarily focus on transactions_df_processed
# and potentially merged data from account_profiles_df_processed.

# Let's ensure the 'is_fraud' column is correctly identified.
# If it's boolean, convert to int for numerical tests if needed, or use directly for grouping.

# Separate fraudulent and legitimate transactions for comparison
legit_txns = transactions_df_processed[transactions_df_processed['is_fraud'] == False]
fraud_txns = transactions_df_processed[transactions_df_processed['is_fraud'] == True]

print(f"Legitimate Transactions: {len(legit_txns)}")
print(f"Fraudulent Transactions: {len(fraud_txns)}")
```

# Summary Statistics: Fraud vs. Legitimate

```python
print("### Summary Statistics for Transaction Amount (Legitimate vs. Fraudulent)")
display(transactions_df_processed.groupby('is_fraud')['amount'].describe())

print("\n### Summary Statistics for IP Risk Score (Legitimate vs. Fraudulent)")
display(transactions_df_processed.groupby('is_fraud')['ip_risk_score'].describe())
```

# Correlation Analysis

```python
print("### Correlation Matrix with Fraud Status")

# Select relevant numerical columns and the target variable
correlation_cols = [
    'amount', 'ip_risk_score', 'velocity_1h', 'account_age_days', 'credit_limit',
    'time_since_last_s', 'amount_vs_avg_ratio', 'is_fraud'
]

# Ensure 'is_fraud' is numeric (0/1) for correlation calculation
transactions_for_corr = transactions_df_processed[correlation_cols].copy()
transactions_for_corr['is_fraud'] = transactions_for_corr['is_fraud'].astype(int)

correlation_matrix = transactions_for_corr.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features and Fraud Status')
plt.show()

print("\n### Correlation of Features with 'is_fraud'")
display(correlation_matrix['is_fraud'].sort_values(ascending=False))
```
